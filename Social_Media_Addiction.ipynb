# Bivariate Analysis

### 1. now we will find out the if there is relation between time spent on social media and relationship status, so we can get to know whether social media is causing realtionship troubles :P
New Section

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='Relationship_Status', y='Avg_Daily_Usage_Hours')
plt.title('Relationship Status vs Time Spent on Social Media')
plt.xlabel('Relationship Status')
plt.ylabel('Time Spent on Social Media')
plt.show()
```

well, from the graph it seems that everyone is having almost same average for time spent on social media! Moreover, "in relationship" status has people with spending more time on social media!! So it social media might be beneficial if your looking for partner :P xD


we dont stop here, we try other methods to conclude if spending time on social media will be helpful for you to get a girlfriend!

```python
group_means = data.groupby('Relationship_Status')['Avg_Daily_Usage_Hours'].mean()
print(group_means)
```

Single people and taken people spend almost same time on social media!!

```python
from scipy.stats import f_oneway
```

now we are performing p-value test to check their relation

```python
group1 = data[data['Relationship_Status'] == 'Single']['Avg_Daily_Usage_Hours']
group2 = data[data['Relationship_Status'] == 'In Relationship']['Avg_Daily_Usage_Hours']
group3 = data[data['Relationship_Status'] == 'Complicated']['Avg_Daily_Usage_Hours']
```

```python
f_stat, p_val = f_oneway(group1, group2, group3)
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")
```

okay, so if p < 0.05, the difference in means is statistically significant and if p >= 0.05, there's no significant difference, and social media time likely doesn't have any relation on relationship status!


So, we found that our p > 0.05 so spending more time doesn't mean that you won't find your partner and it doesn't also mean that you will the partner!

### 2. Now we will see, the more usage of social media has any effects on sleep! Like, people are losing their sleep or not by using more social media!

# Multivariate Analysis

### 1. We will see the combined effect of sleep, mental health and social media addiction on academic performance!

From the graph, we can see that, less sleep and less mental health score is cauisng affects on academic performance which we already observed in bivariate analysis but high social media addiction score is also causing affects on the academic score which was obvious!

### 2. Now we will see which variables are associated with high addiction score!

```python
numeric_data = data[['Addicted_Score', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night',
                     'Mental_Health_Score', 'Conflicts_Over_Social_Media', 'Age']]
```

```python
correlation_matrix = numeric_data.corr()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap with Addicted Score")
plt.show()
```

From the heatmap it is very clear that
* Hours spent of social media is positively correlated with addiction score
* Sleep and mental health score are negatively correlated with addiction score
* Conflicts over social media is positively correlated with addiction score
* Age has very few or no impact or correlation with addiction score


## Observations made from Multivariate Analysis:
1. High addiction score is causing affects on academic performance
2. More hours spent on social media is directly correlated with addiction score which means that more hours spent tends to give high addiction score
3. Sleep has negative correlation with addiction score which means that the less sleep is causing high addiction score
4. Mental health score is also negatively correlated
5. Age has nothing to do with social media addiction. Any person of any age can have addiction!
